Format of target for each task

```python
target_list=['diagnosed_diabetes','diabetes_stage','diabetes_risk_score']
```

Feature Encoding:
Encoding (Text to numeric data) Label Encoding /One-Hot Encoding

```python
import pandas as pd
# 1.Ordinal Mapping
education_map={'No formal':0,'Highschool':1,'Graduate':2,'Postgraduate':3}
income_map={'Low':1,'Lower-Middle':2,'Middle':3,'Upper-Middle':4,'High':5}
diabetes_stage_encoded={'No diabetes':0,'Pre-diabetes':1,'Type 2':2,'Type 1':3}
# Applying mapping
df['education_level']=df['education_level'].str.strip().map(education_map).fillna(0)
df['income_level']=df['income_level'].str.strip().map(income_map).fillna(2)
df['diabetes_stage_encoded']=df['diabetes_stage'].map(diabetes_stage_encoded).fillna(0)
remaining_cats=df.select_dtypes(include=['object']).columns.tolist()
# Applying one-hot encoding
remaining_cats=[c for c in remaining_cats if c not in target_list]
df_final=pd.get_dummies(df,columns=remaining_cats,drop_first=True,dtype=int)
# Checking after encoding
print("first five rows of data")
print(df_final)


```

Converting categorical text data into numeric data because machine only understand numbers .By applying label encoding on these columns (e.g education level) where rank/order is matter change text data into  numbers.And remaining columns of categorical text data where order does not matter convert into numbers (0/1) by applying One Hot Encoding.

```python
# Verification
print("After Encoding")
print("Missing values")
print(df_final.isnull().sum().sum())
df_final.head()
```

Feature and label selection for three tasks  &  Splitting  and Scaling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd

target_list=['diagnosed_diabetes','diabetes_stage','diabetes_risk_score','diebtes_stage_encoded']

# Seperate Features and targets
X=df_final.drop(columns=[c for c in df_final.columns if any (target in c for target in target_list)],errors='ignore')
# For binary classification
y_binary=df_final['diagnosed_diabetes']
# For Multiclass
y_multiclass=df_final['diabetes_stage']
# For Regression
y_regression=df_final['diabetes_risk_score']

# Train_Test Split
X_train,X_test,y_train_bin,y_test_bin,y_train_multi,y_test_multi,y_train_reg,y_test_reg=train_test_split(X,y_binary,y_multiclass,y_regression,stratify=y_binary,test_size=0.2,random_state=42)

# Scaling
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)



print("Splitting and scaling done")
print(f"Original Train Shape:  {X_train_scaled.shape}")






```

In this step ensure features and labels for all three tasks(binary,multiclass and regression) are correctly formated.Splitted data into training and testing into 80% for model training and 20% for testing to evulate model performance what is model behaviour on testing data and to avoid model from overfitting (memorize data) and used stratify to ensure in binary count of target class in train and test remain same  .And scaled data fit(rules learns from only training data) only on training data to avoid data leakage and transfrom on test data.And all tasks model will train on scaled train data.Each task contains it's own target train and test  data but training scaled data which is common features remain same.

1. Binary Classification (Logistic Regression)

```python

import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,recall_score,accuracy_score

# Logistic Regression (Binary Task) Tuning setup
print("Tuning Binary Classification (Logistic Regression)")
param_grid_bin={'C':[0.1,1.0,10],
                'penalty':['l2'],
                'class_weight':['balanced'],
                'max_iter':[1000],
                'solver':['lbfgs']}

# Grid Search with Cross-Validation (CV=5)
grid_search_bin=GridSearchCV(
    estimator=LogisticRegression(random_state=42),
    param_grid=param_grid_bin,
    cv=5,
    scoring='f1',
    verbose=1,
    n_jobs=-1)

# Model fit (Training)
print("Starting Hyperparameter Tuning for Logistic Regression (Binary Classification)")
grid_search_bin.fit(X_train_scaled,y_train_bin)

# Best model selection and updated performance check
best_model_bin=grid_search_bin.best_estimator_

# Prediction for training and testing
y_pred_train_bin=best_model_bin.predict(X_train_scaled)
y_pred_best_bin=best_model_bin.predict(X_test_scaled)

# Accuracy for training and testing
print(f"Training Accuracy Score: {accuracy_score(y_train_bin,y_pred_train_bin):.4f}")
print(f"Testing Accuracy Score: {accuracy_score(y_test_bin,y_pred_best_bin):.4f}")

print("Best Model Parameters:")
print(best_model_bin.get_params())
# Results
print(f"Classification report{classification_report(y_test_bin,y_pred_best_bin)}")
print(f"Updated Recall Score {recall_score(y_test_bin,y_pred_best_bin):.4f} ")
print(f"Final accuracy score {accuracy_score(y_test_bin,y_pred_best_bin):.4f}")




```
